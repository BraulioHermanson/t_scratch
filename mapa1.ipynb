```python
## adicionando Marcadores
mapa = folium.Map(location = [-27.592704,-48.5045356],
           tiles = 'Stamen Terrain',
           zoom_start=10)
```

```python
folium.Marker(
    [-27.592704,-48.5045356],
    popup='<i>UFSC - Universidade Federal de Santa Catarina<i/>',
    tooltip = 'Clique aqui',
    icon=folium.Icon(color='red')
).add_to(mapa)
```

```python
folium.CircleMarker(
    location=[-27.592704,-48.5045356],
    radius=50,
    color='#3186cc',
    fill=True,
    fill_color='31186cc'
).add_to(mapa)
```

```python
df = pd.read_excel('arquivo.xlsx')
```

```python
df.info()
```

```python
coordenadas=[]
for lat, lng in zip(df.Latitude.values,df.Longitude.values):
    coordenadas.append([lat,lng])
```

```python
mapa = folium.Map(location=[-27.592704,-48.5045356],
                  tiles = 'Stamen Terrain',
                  zoom_start=5
                  )
```

```python
mapa.add_child(plugins.HeatMap(coordenadas))
mapa
```
